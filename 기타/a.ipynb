```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
# 자유도 설정
df_values = [1, 2, 3, 5, 10, 20, 50]

# x 값 생성
x = np.linspace(0, 30, 1000)

# 그래프 설정
plt.figure(figsize=(12, 8))

# 각 자유도에 대해 카이제곱 분포의 PDF를 계산하고 플로팅
for df in df_values:
    y = chi2.pdf(x, df)
    plt.plot(x, y, label=f'df={df}')

plt.title('카이제곱 분포 (자유도에 따른 차이)')
plt.xlabel('x')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# 자유도 설정
df_values = [1, 2, 5, 10, 20]

# x 값 생성
x = np.linspace(-5, 5, 1000)

# 그래프 설정
plt.figure(figsize=(12, 8))

# 각 자유도에 대해 t-분포의 PDF를 계산하고 플로팅
for df in df_values:
    y = t.pdf(x, df)
    plt.plot(x, y, label=f'df={df}')

# 표준정규분포 추가
y_norm = t.pdf(x, np.inf)
plt.plot(x, y_norm, label='Normal distribution', linestyle='--', color='black')

plt.title('t-분포 (자유도에 따른 차이)')
plt.xlabel('x')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# 자유도 설정
dfn_values = [1, 2, 5, 10]  # 분자 자유도
dfd_values = [1, 2, 5, 10]  # 분모 자유도

# x 값 생성
x = np.linspace(0, 5, 1000)

# 그래프 설정
plt.figure(figsize=(12, 8))

# 각 자유도 조합에 대해 F-분포의 PDF를 계산하고 플로팅
for dfn in dfn_values:
    for dfd in dfd_values:
        y = f.pdf(x, dfn, dfd)
        plt.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')

plt.title('F-분포 (자유도에 따른 차이)')
plt.xlabel('x')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```
