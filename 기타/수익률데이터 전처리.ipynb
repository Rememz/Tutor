```python
import pandas as pd
import os

# 데이터가 저장된 폴더 경로
folder_path = './data'

# 폴더 내 모든 파일 리스트 가져오기
file_list = os.listdir(folder_path)
file_list = file_list
# xlsx 파일만 필터링
xlsx_files = [file for file in file_list if file.endswith('.xlsx')]

# 데이터프레임을 저장할 딕셔너리
dataframes = {}

# 각 xlsx 파일을 데이터프레임으로 읽어서 딕셔너리에 저장
for file in xlsx_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path,header=3)
    dataframes[file] = df

# 각 데이터프레임에 접근하려면
# 예: 첫 번째 xlsx 파일의 데이터프레임
first_df = dataframes[xlsx_files[0]]
print(first_df.head())
```
