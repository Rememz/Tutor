```python
#변수선택
from sklearn.feature_selection import SelectKBest, chi2

num = ["총매출액", "1회 평균매출액", "할인권 사용 횟수", "총 할인 금액", "구매카테고리수", "Frequency", "Monetary"]
cg = ["구매금액대", "고객등급", "구매유형", "클레임접수여부", "거주지역", "성별", "고객 나이대"]

X = df.drop(["고객ID", "이탈여부"], axis = 1)
Y = df["이탈여부"]

#분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

#데이터 밸런싱
# X_train, Y_train = SMOTE(random_state=0).fit_resample(X_train, Y_train)

#일변량 통계기반
selector = SelectKBest()
selector.fit(X_train, Y_train)

score = pd.DataFrame()
score["Chi2"] = selector.scores_ #chi2 = 카이제곱 값
score["p-value"] = selector.pvalues_
score["support"] = selector.get_support()
score["attribute"] = X_train.columns
score[score["support"]==True] #support True인것들만 뽑아주기
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier #앙상블

selector = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=0), threshold="median") #threshold="median" 상위 50% 이상의 모델들 부터 선택
selector.fit(X_train, Y_train)

score = pd.DataFrame()
score["attribute"] = X_train.columns
score["support"] = selector.get_support()
score[score["support"]==True]
```

```python
from pandas.core.common import random_state
from sklearn.linear_model import LogisticRegression

logit =LogisticRegression(C=1, random_state=0)

logit.fit(X_train, Y_train)
Y_pred = logit.predict(X_test)

#Y_pred
print("accuracy(R2) : {:.3f}".format(logit.score(X_train, Y_train)))

from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
print(f"intercept : {logit.intercept_}")
print(f" beta : {logit.coef_}")
```
