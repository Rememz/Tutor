```python
import pandas as pd
import numpy as np  
```

```python
train = pd.read_csv("./데이터/버스 운행시간 데이터/train.csv")
test = pd.read_csv("./데이터/버스 운행시간 데이터/test.csv")
submission = pd.read_csv("./데이터/버스 운행시간 데이터/submission_제출양식.csv")
```

```python
train.drop('id',axis=1,inplace=True)
test.drop('id',axis=1,inplace=True)
```

```python
train
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoder.fit(train["next_station"])
```

```python
train["now_station"] = encoder.transform(train["now_station"])
train["next_station"] = encoder.transform(train["next_station"])
```

```python
test["now_station"] = encoder.transform(test["now_station"])
test["next_station"] = encoder.transform(test["next_station"])
```

```python
train
```

```python
# date 의 object 타입을 Datetime으로 변경
train["date"]=pd.to_datetime(train["date"])
test["date"]=pd.to_datetime(test["date"])
```

```python
train.drop(["now_latitude","now_longitude","next_latitude","next_longitude"],axis=1, inplace=True)
test.drop(["now_latitude","now_longitude","next_latitude","next_longitude"],axis=1, inplace=True)
```

```python
# 0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일
train["weekday"]=train["date"].dt.weekday
test["weekday"]=test["date"].dt.weekday
train
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

encoder.fit(train["route_id"])
```

```python
train["route_id"] = encoder.transform(train["route_id"])
test["route_id"] = encoder.transform(test["route_id"])
```

```python
train.drop(["date","vh_id","route_nm"],axis=1,inplace=True)
test.drop(["date","vh_id","route_nm"],axis=1,inplace=True)
```

```python
list = train.groupby(["now_arrive_time"])["distance"].mean().sort_values(ascending=False).index.tolist()
list
```

```python
a = 0
for i in list:
    train["now_arrive_time"] = train.now_arrive_time.replace(i,a)
    test["now_arrive_time"] = test.now_arrive_time.replace(i,a)
    a += 1
    
    
```

```python
train.drop(["now_station","next_station"],axis=1,inplace=True)
test.drop(["now_station","next_station"],axis=1,inplace=True)
```

```python
feature = train.drop("next_arrive_time",axis=1)
target = train["next_arrive_time"]
```

```python
import statsmodels.api as sm

feature_add = sm.add_constant(feature,  has_constant='add')

# sm OLS 적합
model = sm.OLS(target , feature_add)
fitted_model = model.fit()

# summary 함수통해 결과출력
fitted_model.summary()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(feature, target)
```

```python
pred = lr.predict(test)
pred
```

```python
submission2 = submission
submission2["next_arrive_time"] = pred
```
