```python
test["now_arrive_time"]=test["now_arrive_time"].str.replace('시','')
test["now_arrive_time"]=test["now_arrive_time"].astype(int)
test["route_nm"]=test["route_nm"].astype(int)
```

```python
train.drop(["now_arrive_time"],axis=1,inplace=True)
test.drop(["now_arrive_time"],axis=1,inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
```

```python
feature = train.drop("next_arrive_time",axis=1)
target = train["next_arrive_time"]
```

```python
from sklearn.preprocessing import MinMaxScaler

# 변형 객체 생성
minmax_scaler = MinMaxScaler()

# 훈련데이터의 모수 분포 저장
minmax_scaler.fit(feature)

# 훈련 데이터 스케일링
feature_scaled = minmax_scaler.transform(feature)

# 테스트 데이터의 스케일링
test_scaled = minmax_scaler.transform(test)

# 스케일링 된 결과 값으로 본래 값을 구할 수도 있다.
# X_origin = minmax_scaler.inverse_transform(feature_scaled)

feature_scaled = pd.DataFrame(data = feature_scaled, columns=feature.columns)
test_scaled = pd.DataFrame(data = test_scaled, columns=test.columns)
# y_train = pd.DataFrame(y_train)
# y_train.reset_index(drop='index', inplace=True)
# y_test = pd.DataFrame(y_test)
# y_test.reset_index(drop='index', inplace=True)
```

```python
import statsmodels.api as sm

feature_add = sm.add_constant(feature_scaled,  has_constant='add')

# sm OLS 적합
model = sm.OLS(target , feature_add)
fitted_model = model.fit()

# summary 함수통해 결과출력
fitted_model.summary()
```

```python
test_add = sm.add_constant(test_scaled,  has_constant='add')
```

```python
lr.fit(feature_add, target)
```

```python
pred = lr.predict(test_add)
pred
```

```python
submission2 = submission
submission2["next_arrive_time"] = pred
```
