파이썬에서 Set(집합)은 중복되지 않는 요소들로 구성된 변경 가능한 자료형입니다.\
Set은 중괄호 {}를 사용하여 생성하며, 각 요소는 쉼표로 구분됩니다. Set은 순서가 없기 때문에 인덱스를 사용하여 요소에 접근할 수 없습니다.

Set은 다음과 같은 특징을 가지고 있습니다:

중복된 요소를 허용하지 않습니다: Set은 중복된 값을 허용하지 않습니다. \즉, Set에는 동일한 값을 여러 번 포함할 수 없습니다. 중복된 값을 제거하고 고유한 요소들만 남기는 용도로 자주 사용됩니다.

요소들의 순서가 없습니다: Set은 요소들이 순서 없이 저장됩니다.\ 따라서 인덱스를 사용하여 특정 요소에 직접 접근할 수 없습니다. 순서가 중요하지 않고 고유한 요소들의 집합을 다룰 때 유용합니다.

Set은 다양한 연산을 지원합니다. 일반적인 Set 연산에는 교집합, 합집합, 차집합, 대칭 차집합 등이 있습니다. \
예를 들어, 두 개의 Set A와 B가 있다고 가정하면 다음과 같은 연산을 수행할 수 있습니다:

교집합: A.intersection(B) 또는 A & B를 사용하여 두 Set의 교집합을 구할 수 있습니다.\
합집합: A.union(B) 또는 A | B를 사용하여 두 Set의 합집합을 구할 수 있습니다.\
차집합: A.difference(B) 또는 A - B를 사용하여 첫 번째 Set에서 두 번째 Set을 제외한 차집합을 구할 수 있습니다.\
대칭 차집합: A.symmetric_difference(B) 또는 A ^ B를 사용하여 두 Set의 대칭 차집합을 구할 수 있습니다.\
Set은 추가, 삭제, 갱신 등의 연산도 지원합니다.\ 예를 들어, 요소를 Set에 추가하려면 add() 메서드를 사용하고, 요소를 제거하려면 remove() 메서드를 사용합니다.

```python
# 리스트에서 2가지 값중 공통된 값찾기
def find_common_elements(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return list(set1.intersection(set2))
```
